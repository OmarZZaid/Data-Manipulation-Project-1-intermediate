````python
Energy.rename(columns={'Energy Supply':'Energy_Supply','Energy Supply per Capita':'Energy_Supply_per_Capita','% Renewable':'Renewable%'}, inplace=True)
````

````python
def keep_number(s):
    s2 = ''.join([i for i in s if i.isdigit()])
    return s2
````

````python
def remove_digits(s):
    new_string = ''
    for character in s:
        if not character.isdigit():
            new_string += character
    return new_string
````

````python
def remove_brackets(s):
    new_string = ''
    for i in s:
        if i == '(':
            new_string= s[:s.index('(')] + s[s.index(')')+1:]
            return new_string
    return s
````

````python
Energy['Energy_Supply'],Energy['Energy_Supply_per_Capita']= Energy['Energy_Supply'].astype(str),Energy['Energy_Supply_per_Capita'].astype(str)
````

````python
Energy['Energy_Supply']=Energy['Energy_Supply'].apply(keep_number)
Energy['Energy_Supply_per_Capita']=Energy['Energy_Supply_per_Capita'].apply(keep_number)
Energy['Country']=Energy['Country'].apply(remove_digits)
Energy['Country']=Energy['Country'].apply(remove_brackets)
Energy['Energy_Supply_per_Capita']=Energy['Energy_Supply_per_Capita'].apply(keep_number)
````

````python
Energy['Energy_Supply']= pd.to_numeric(Energy['Energy_Supply'])
Energy['Energy_Supply_per_Capita']= pd.to_numeric(Energy['Energy_Supply_per_Capita'])
````

````python
Energy['Energy_Supply']=Energy['Energy_Supply']*1000000
````
